Load your credentials file

```python
credentials = yaml.load(open('credentials.yml'))
```

Load your individual credentials

```python
twitter_key = credentials['twitter_key']
twitter_secret = credentials['twitter_secret']
token = credentials['twitter_token']
token_secret = credentials['twitter_tokensecret']
```

Authenticate with twitter API

```python
import twitter
```

```python
auth = twitter.oauth.OAuth(token, token_secret, 
                           twitter_key, twitter_secret)
twitter_api = twitter.Twitter(auth=auth)
```

Request data from Twitter

```python
search_string = '#trump' 
count = 100
search_results = twitter_api.search.tweets(q=search_string, count=count)
```

Create lists for the t
ext, created date and name for all the tweets

```python
texts = []
created = []
name = []

for n in range(len(search_results['statuses'])):
    texts.append(search_results['statuses'][n]['text'])
    created.append(search_results['statuses'][n]['created_at'])
    name.append(search_results['statuses'][n]['user']['name'])
```

Build a DataFrame with the lists

```python
import pandas as pd
```

```python
df = pd.DataFrame({'text':texts,'created_at':created, 'name':name})
df
```
