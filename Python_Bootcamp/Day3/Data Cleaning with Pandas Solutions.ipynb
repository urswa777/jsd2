Import pandas

```python
import pandas as pd
```

Read the combine.csv file as a DataFrame and investigate the contents

```python
data = pd.read_csv('combine.csv')
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
print data.head()
print data.info()
print data.describe()
```

Fill missing values for college with 'No College'

```python
print data.info()
data.college.fillna('No College',inplace=True)
print data.info()
```

Remove the players that have null values for the pick

```python
print data.info()
data_dropped = data.dropna(subset=['pick'])
print data_dropped.info()
```

Investigate the unique values in the position column

```python
data.position.unique()
```

Replace RB and QB with Running Back and Quarterback

```python
data.position.replace(['RB','QB'],['Running Back','Quarterback'],inplace=True)
data.head(20)
```

Create dummy values for position

```python
data_positions = pd.get_dummies(data.position,prefix='Pos')
data_positions.head()
```

Merge the dummy data with the original data set

```python
data = pd.merge(data, data_positions, left_index=True, right_index=True)
data.head()
```

Convert weight from lbs to kg

```python
data['weight_kg'] = data.weight/2.2
data.head()
```

Capitalize name

```python
data.name.apply(str.upper).head()
```

Reverse order of first and last name

```python
data.apply(lambda x:'{0}, {1}'.format(
        x['lastname'], x['firstname']), 
        axis=1).head()
```

```python
def reverse_name(name):
    return "{0}, {1}".format(name['lastname'],name['firstname'])
```

```python
data.apply(reverse_name,axis=1)
```
