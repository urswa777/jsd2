Import your packages

```python
import seaborn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Add "magic command" for iPython Notebook to display plots in the notebook

```python
%matplotlib inline
```

Create mock data for plots

```python
x = np.arange(20)
y = np.random.normal(10,1,20)
z = np.random.normal(10,2,20)
```

Create your first plot

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x,y)
```

Create two plots with line and scatterplot

```python
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax1.plot(x,y)
ax2 = fig.add_subplot(122)
ax2.scatter(y,z)
```

Create two plots with histogram and boxplot

```python
fig = plt.figure()
ax3 = fig.add_subplot(121)
ax3.hist(y)
ax4 = fig.add_subplot(122)
ax4.boxplot([y,z])
```

Plot multiple data sets on same axes with labels and a legend

```python
fig = plt.figure()
ax5 = fig.add_subplot(111)
ax5.plot(x,y,label = 'Thing #1')
ax5.plot(x,z,label = 'Thing #2')
ax5.legend()
```

Add titles and suptitle to figure 

```python
fig = plt.figure()
ax7 = fig.add_subplot(111)
ax7.plot(x,y)
ax7.set_title("Line")
ax8 = fig.add_subplot(122)
ax7.scatter(y,z)
ax7.set_title("Scatter")
fig.suptitle('Two Charts')
```

```python
fig = plt.figure()
ax7 = fig.add_subplot(211)
ax7.plot(x,y)
ax7.set_title("Line")
ax8 = fig.add_subplot(222)
ax7.scatter(y,z)
ax7.set_title("Scatter")
fig.suptitle('Two Charts')
```

Add x and y axis labels with font sizes

```python
fig = plt.figure()
ax8 = fig.add_subplot(111)
ax8.plot(x,y)
ax8.set_ylabel('Y-Axis', fontsize=10)
ax8.set_xlabel('X-Axis', fontsize=10)
```

Save your figure as png file

```python
fig.savefig('figure.png')
```
