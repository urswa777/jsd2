Import your packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

Add "magic command" for iPython Notebook to display plots in the notebook

```python
%matplotlib inline
```

Create mock data for plots

```python
x = np.arange(20)
y = np.random.normal(10,1,20)
z = np.random.normal(10,2,20)
```

Create your first plot

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x,y)
```

Create two plots with line and scatterplot

```python
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax1.plot(x,y)
ax2 = fig.add_subplot(122)
ax2.scatter(y,z)
```

Create two plots with histogram and boxplot

```python
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.hist(y)
ax2.boxplot([y,z]);
```

Plot multiple data sets on same axes with labels and a legend

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y, label = 'Thing #1')
ax.plot(x, z, label = 'Thing #2')
ax.legend();
```

Add titles and suptitle to figure 

```python
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax1.plot(x,y)
ax1.set_title("Line")
ax2 = fig.add_subplot(122)
ax2.scatter(y,z)
ax2.set_title("Scatter")
fig.suptitle('Two Charts')
```

Add x and y axis labels with font sizes

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y)
ax.set_ylabel('Y-Axis', fontsize=10)
ax.set_xlabel('X-Axis', fontsize=20);
```

Save your figure as png file

```python
fig.savefig('figure.png')
```

## Exercise

#### Create a figure with 4 subplots
- Histogram of Titanic age distribution
- Scatterplot of Age and Ticket Fares
- Boxplot of Ages for Survivors vs. Deceased
- Boxplot of Ages for Men vs. Women
- Save figure as a png file


```python
# Clean data for charts
data = pd.read_csv('clean_data.csv')

print data.head()
print data.info()
print data.describe().T
```

```python
# Histogram data
hist_data = data.Age

# Scatter plot data
scatter_age = data.Age
scatter_fare = data.Fare

# Box plot data for age of survivors vs. deceased
age_survived = data.Age[data.Survived==1]
age_deceased = data.Age[data.Survived==0]

# Box plot data for age of men vs. women
age_men = data.Age[data.Sex==1]
age_women = data.Age[data.Sex==0]
```

```python
fig = plt.figure()

# Histogram of passenger ages
ax1 = fig.add_subplot(221)
ax1.hist(hist_data, bins=30)
ax1.set_xlabel('Age in Years')
ax1.set_ylabel('Number of Passengers')
ax1.set_title('Titanic Passenger Ages', fontsize=8)

# Scatter plot
ax2 = fig.add_subplot(222)
ax2.scatter(scatter_age,scatter_fare, 
            s=5, #size
            c='#4093DD', #color
            alpha=.2, #transparency
            marker='o')#marker symbol
ax2.set_xlabel('Age in Years')
ax2.set_ylabel('Ticket Fare')
ax2.set_title('Ticket Fare vs. Age', fontsize=8)
ax2.set_ylim(0,200)

# Box plot for age of survisors vs. deceased
ax3 = fig.add_subplot(223)
ax3.boxplot([age_survived, age_deceased])
ax3.set_ylabel('Age in Years')
ax3.set_xticklabels(['Survivors','Deceased'])
ax3.set_title('Titanic Passenger Ages', fontsize=8)

# Box plot  for age of men vs. women
ax4 = fig.add_subplot(224)
ax4.boxplot([age_men, age_women])
ax4.set_ylabel('Age in Years')
ax4.set_xticklabels(['Men','Women'])
ax4.set_title('Titanic Passenger Ages', fontsize=8)
plt.tight_layout()
plt.savefig('Matplotlib_Example.png')
```

Pie Chart Example

```python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = [0, 0, 0, 0.2] # only "explode" the 2nd slice (i.e. 'Hogs')

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
# Set aspect ratio to be equal so that pie is drawn as a circle.
plt.axis('equal');

# plt.show()
```
